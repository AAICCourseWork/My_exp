```python
from tqdm import tqdm

data = pd.read_csv("5_a.csv") 
#print(data['proba'])
y_pred=[]
#condition for each row in dataframe
for ind in data.index:
    if data['proba'][ind]<0.5:
        y_pred.append(0)
    else:
        y_pred.append(1)
#print(y_pred)
#confusion matrix
labelcount=2#since only two labels 0/1
TN=0
FN=0
FP=0
TP=0
for ind in data.index:
    if (data['y'][ind]==1 and y_pred[ind]==1):
        TP=TP+1
    elif data['y'][ind]==1 and y_pred[ind]==0:
        FN=FN+1
    elif data['y'][ind]==0 and y_pred[ind]==0:
        TN=TN+1
    elif data['y'][ind]==0 and y_pred[ind]==1:
        FP=FP+1
#print(TP)
#print(FN)
#print(TN)
#print(FP)
print("confusion matrix\n[[{0},{1}]\n[{2},{3}]]".format(TN,FN,FP,TP))
P=FN+TP#total positive points
precision=(TP/(TP+FP))
recall=(TP/P)
F1_score=2*((precision*recall)/(precision+recall))
print("F1 score = {}".format(F1_score))
accuracy_score=((TP+TN)/(TP+FN+TN+FP))
#calculating AUC

#print(proba_sort)
TPR=[]
FPR=[]

sorted_desc = data.sort_values('proba',ascending = False)
tp_i=tn_i=fp_i=fn_i=0
TPR_i=FPR_i=0
N_i = data['y'].value_counts()[0] 
P_i= data['y'].value_counts()[1] 

for i in tqdm(sorted_desc.proba):
    Threshold_pred=[]
    for ind in sorted_desc.proba:
        if ind<i:
            auc=0
            Threshold_pred.append(auc)
            
        else:
            auc=1
            Threshold_pred.append(auc)
    sorted_desc['threshold'] = Threshold_pred 

    for k in data:
        TP = ((sorted_desc['y']==1.0) & (sorted_desc['threshold'] == 1.0)).sum()   
        FP = ((sorted_desc['y']==0.0) & (sorted_desc['threshold'] == 1.0)).sum()
    TPR_i = TP/P_i
    FPR_i = FP/N_i
    TPR.append(TPR_i)
    FPR.append(FPR_i)
auc1 = np.trapz(TPR,FPR)
print("auc score ",auc1)
print("accuracy score",accuracy_score)


 
    


            
        
```
