<h2><font color='Geen'>  Implementing Decision Function of SVM RBF Kernel</font></h2>

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
import numpy as np
from sklearn.svm import SVC
```

```python
X, y = make_classification(n_samples=5000, n_features=5, n_redundant=2,
                           n_classes=2, weights=[0.7], class_sep=0.7, random_state=15)
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
X_train, X_cv, y_train, y_cv = train_test_split(X_train, y_train, test_size=0.250, stratify=y_train)
```

```python
print(X_test.shape)
print(X_cv.shape)
X_train.shape
```

```python
from sklearn.svm import SVC
clf = SVC(gamma=0.001, C=100)
clf.fit(X_train, y_train)
```

```python
from numpy import linalg as LA
import math
def decision1_function(Xcv):
  dc=[]
  gamma=0.001
  sum=0
  alpha=clf.dual_coef_[0]
  for Xq in Xcv:
    a=0
    for i in range(len(clf.support_vectors_)):
      norm=LA.norm(clf.support_vectors_[i]-Xq)
      K=math.exp(-gamma*(norm**2))
     
      a=a+(alpha[i]*K)

    final=a+clf.intercept_
    #print(a)
    dc.append(final)
  return dc
```

```python
dm=decision1_function(X_cv)
fcv=[]
for i in dm:
  fcv.append(list(i))
#print("decision function of cv",fcv)

ds=decision1_function(X_test)
ftest=[]
for i in ds:
  ftest.append(list(i))
#print("decision function of test",ftest)
```

```python
fcv_cal=[]
for i in fcv:
  fcv_cal.append(i[0])
d=np.array(fcv_cal)
print("custom implementation -decision function of cv",d)


```

```python
print("clf decision function of cv", clf.decision_function(X_cv))
```

```python
ftest_cal=[]
for i in ftest:
  ftest_cal.append(i[0])
m=np.array(ftest_cal)
print("custom implementation-decision function of test",m)
```

```python
print("clf decision function of test",clf.decision_function(X_test))
```

<h2><font color='Geen'> Implementing Platt Scaling to find P(Y==1|X)</font></h2>

```python
#print(y_cv)
Npos=0
Nneg=0
for i in y_train:
  if i==1:
    Npos+=1
  else:
    Nneg+=1
Ypos=((Npos+1)/(Npos+2))
Yneg=(1/(Nneg+2))
#print(Ypos)
#print(Yneg)
y_cv_cal=[]
for i in y_cv:
  if i==1:
    y_cv_cal.append(Ypos)
  else:
    y_cv_cal.append(Yneg)
#print(y_cv_cal)
```

```python
#print(y_cv)
Npos=0
Nneg=0
for i in y_test:
  if i==1:
    Npos+=1
  else:
    Nneg+=1
Ypos_test=((Npos+1)/(Npos+2))
Yneg_test=(1/(Nneg+2))
#print(Ypos)
#print(Yneg)
y_test_cal=[]
for i in y_test:
  if i==1:
    y_test_cal.append(Ypos_test)
  else:
    y_test_cal.append(Yneg_test)
#print(y_test_cal)
```

```python

from sklearn.metrics import log_loss
w = np.zeros_like(fcv_cal[0])
b = 0
eta0  = 0.0001
alpha = 0.0001
N = len(fcv_cal)
M=len(X_test)

def sigmod(x):
    if x<0:
        return 1-1/(1+math.exp(x))
    else:
        return 1/(1+math.exp(-x))
#print(len(inti_loss_train))
#print(prob[0])
prob1=[]
inti_loss_train1=[]
prob2=[]
inti_loss_test=[]
#print(len(y_train))

for i in range(N):
    #print(wt)
    #print(b)
    prob1.append(np.dot(w,fcv_cal[i])+b)
#print(len(prob1))
for i in range(len(prob1)):
    #print(prob[i])
    inti_loss_train1.append(1 / (1 + sigmod(-prob1[i])))
#print(len(inti_loss_train1))
#print(log_loss(y_train,inti_loss_train1,eps=0.0001))


k=0
for i in  range(N):
    k=k+(np.log(inti_loss_train1[i])*y_cv_cal[i])+(np.log(1-inti_loss_train1[i]))*(1-y_cv_cal[i])
log_loss_trian=-(k/N)
print("initial train loss",log_loss_trian)  


#calculating test set log loss
for i in range(M):
    prob2.append(np.dot(w,ftest_cal[i])+b)
#print(prob[0])
for i in range(len(prob2)):
    
    inti_loss_test.append(1 / (1 + sigmod(-prob2[i])))
#print(len(inti_loss_train))
k=0
for i in  range(M):
    k=k+(np.log(inti_loss_test[i])*y_test_cal[i])+((np.log(1-inti_loss_test[i]))*(1-y_test_cal[i]))
log_loss_test=-(k/M)
print("initial test loss",log_loss_test)
#print(log_loss(y_test,inti_loss_test,eps=0.0001))
```
